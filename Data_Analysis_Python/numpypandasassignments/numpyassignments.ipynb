```python
# Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.

import numpy as np

array = np.random.randint(1,21,size = (5,5))

print("Original Array")

print(array)

print("Modified Array")

array[:,2] = 1

print(array)
```

```python
# Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.

arr = np.arange(1,17).reshape(4,4)

print("Original Array")

print(arr)

print("Modified Array")

np.fill_diagonal(arr,0)

print(arr)
```

```python
## Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.

arr = np.arange(1,37).reshape(6,6)

print("Original Array")

print(arr)

print("Extracted Rows and Columns")

print(arr[2:5,1:4])
```

```python
## Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

arr = np.arange(1,26).reshape(5,5)

print("Original Array")

print(arr)

print("Extracted Border elements")

border_elements = np.concatenate((arr[0, :], arr[-1, :], arr[1:-1, 0], arr[1:-1, -1]))
print("Border elements:")
print(border_elements)

```

```python
## Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.

array_1 = np.random.randint(1,11,size=(3,4))
array_2 = np.random.randint(1,11,size=(3,4))
print("Array1")
print(array_1)
print("Array2")
print(array_2)

addition = array_1+array_2

substraction = array_1-array_2

Multiplication = array_1*array_2

division = array_1/array_2

print("elements wise addition")
print(addition)
print("elements wise substraction")
print(substraction)
print("elements wise multiplication")
print(Multiplication)
print("elements wise division")
print(division)
```

```python
## Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

arr = np.arange(1,17).reshape(4,4)

print("Array")

print(arr)

print("Compute - row - wise")

row_sum = np.sum(arr,axis=1)

print(row_sum)

print("Compute - column - wise")

column_sum = np.sum(arr,axis=0)

print(column_sum)



```

```python
## Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

arr = np.random.randint(1,21,size=(5,5))

print(arr)

mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)
var = np.var(arr)

print("Mean:",mean)
print("Median:",median)
print("Standard deviation:",std_dev)
print("Variance:",var)
```

```python
## Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

arr = np.arange(1,10).reshape(3,3)

print(arr)

# Normalization

mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)
normalization = (arr-mean)/std_dev

print("Normalized Array:")
print(normalization)
```
