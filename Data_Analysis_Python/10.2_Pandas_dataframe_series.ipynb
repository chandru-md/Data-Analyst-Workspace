```python
## Pandas is a poweful data manipulation library in Python, widely used for data analysis and data cleaning.

## It provides two primary data structures: Series and Data frame.

## A Series is a one- dimensional array like object, while a Data frame is a two dimensional, size-mutable and potentially heterogeneous tabular data structure with labeled axes
## (rows and columns).
```

```python
import pandas as pd
```

```python
## Series

## A Pandas - Series is one dimensional array - like object that can hold any data type.
## It is similar to  column in a table.
```

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series:\n",series)
print(type(series))
```

```python
## Create series from ditionary

data = {'a':1,'b':2,'c':3,'d':4,'e':5}

series_dict = pd.Series(data)

print("Series_dict:\n",series_dict)
print(type(series_dict))
```

```python
data = [10,20,30]

index = ['a','b','c']

s = pd.Series(data,index=index)

print(type(s))
```

```python
## Data Frame
```

```python
## Create a data frame from a dictionary of list

data = {
    "Name":["Chandru","Lokesh","Hari Krishna"],
    "Age":[24,25,24],
    "Location":["IND","USA",'UK']
}

df = pd.DataFrame(data)

print(df)

print(type(df))
```

```python
## Create a data frame from a list of dictionaries

data = [

    {"Name":"Chandru","Age":24,"Location":"IND"},
    {"Name":"Hari Krishna","Age":25,"Location":"UK"},
    {"Name":"Lokesh","Age":26,"Location":"USA"}

]

df = pd.DataFrame(data)

print(df)
```
