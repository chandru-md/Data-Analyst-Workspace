```python
## understanding the variable types

## python is dynammically typed in python, type of a variable is determined at run time

name = "Chandru"
age = 24
height = 5.8
is_founder = False

print(type(name))
print(type(age))
print(type(height))
print(type(is_founder))
```

```python
## type checking and conversion = type casting

print(type(age))

## type conversion

age_str = str(age)

print(age_str)

print(type(age_str))
```

```python
# Simple Calculator using variables

# Step 1: Store numbers in variables
num1 = float(input("Enter your first number: "))  # first number
num2 = float(input("Enter your second number: "))    # second number

# Step 2: Perform operations using variables
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2

# Step 3: Print results
print("First number:", num1)
print("Second number:", num2)
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```
